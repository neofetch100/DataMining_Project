Data Cleaning and Transform



```python
def clean_data(df):

    df = df.drop_duplicates()

    # numeric ma convert
    df["Age"] = pd.to_numeric(df["Age"], errors="coerce")
    df["Purchase_Amount"] = pd.to_numeric(df["Purchase_Amount"], errors="coerce")

    # Handle missing numerical values
    df["Age"] = df["Age"].fillna(df["Age"].median())
    df["Purchase_Amount"] = df["Purchase_Amount"].fillna(df["Purchase_Amount"].median())

    # Handle unrealistic age values
    df.loc[(df["Age"] < 0) | (df["Age"] > 100), "Age"] = df["Age"].median()

    # Handle missing categorical values
    df["Product_Category"] = df["Product_Category"].fillna(df["Product_Category"].mode()[0])
    df["Payment_Method"] = df["Payment_Method"].fillna(df["Payment_Method"].mode()[0])

    # Clean Gender properly
    df["Gender"] = df["Gender"].astype(str).str.strip().str.lower()
    df["Gender"] = df["Gender"].replace({
        "male": "Male",
        "m": "Male",
        "female": "Female",
        "f": "Female"
    })

    # Standardize other categorical columns
    df["Product_Category"] = df["Product_Category"].astype(str).str.strip().str.capitalize()
    df["Payment_Method"] = df["Payment_Method"].astype(str).str.strip().str.capitalize()

    # Convert types
    df["Age"] = df["Age"].astype(int)
    df["Purchase_Amount"] = df["Purchase_Amount"].astype(float)

    return df
```

```python
csv_cleaned = clean_data(csv_data)
xlsx_cleaned = clean_data(xlsx_data)
json_cleaned = clean_data(json_data)
```

```python
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import LabelEncoder
```

```python
scaler = MinMaxScaler()
le = LabelEncoder()
def transfrom_data(df):
    df[["Purchase_Amount"]] = scaler.fit_transform(
        df[["Purchase_Amount"]]
    )
    df["Gender"] = le.fit_transform(df["Gender"])

    df["Age_Group"] = pd.cut(
    df["Age"],
    bins=[0, 18, 30, 50, 100],
    labels=["Teen", "Young", "Adult", "Senior"]
)

    return df
```

```python
csv_transformed = transfrom_data(csv_cleaned)
xlsx_transformed = transfrom_data(xlsx_cleaned)
jsn_transformed = transfrom_data(json_cleaned)
```
